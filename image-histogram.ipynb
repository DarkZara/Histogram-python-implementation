```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
```

```python
def histogram(image):
  #histogram is a 0-255 array presenting the number of certain brightness occurances in image
  hist = np.zeros(256,dtype=int)
  #to speed up this process and avoid multi-looping we are going to find unique vaalues of image and their counts
  uni, cnt = np.unique(image, return_counts=True)
  # and pass these information to histogram
  for i,j in zip(uni,cnt):
    hist[uni[i]]=j
  hist=hist.astype(int)
  return hist
```

```python
image=cv2.imread("img1.jpg",cv2.IMREAD_GRAYSCALE)
#everything is clear
cv_hist = cv2.calcHist(image,[0],None,[256],[0,256])
cv_hist2=histogram(image)

plt.imshow(image,cmap='gray')
plt.title('the image')
plt.show()

plt.plot(cv_hist)
plt.title('Histogram for gray scale picture by cv2')
plt.show()

plt.plot(cv_hist2)
plt.title('Histogram for gray scale picture by func')
plt.show()
```
